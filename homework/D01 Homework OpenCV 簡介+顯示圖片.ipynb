思考一下我們前面有提到圖片是矩陣，但維度可能會不一樣
例如灰階圖只有兩個維度，RGB 彩圖則有 3 個維度

假如今天我們把 RGB 3 個維度拆開來看會有甚麼不同的效果呢？

```python
import cv2
import matplotlib.pyplot as plt

plt.figure(figsize = (10,10))
for i,x in enumerate({'hsv' : cv2.IMREAD_COLOR,
                      'gray' : cv2.IMREAD_GRAYSCALE}.items()):
    img = cv2.imread('.\D00_lena.png', x[1])

#     # opencv 的開啟圖片
#     cv2.imshow(x[0],img)
#     cv2.waitKey(0)
#     cv2.destroyAllWindows()

    # matplotlib 的開啟圖片
    plt.subplot(1, 2, i+1, title = x[0])
    plt.imshow(img[:, :, [2, 1, 0]] if x[0] == 'hsv' else img, cmap = x[0])
    plt.axis('off')
plt.show()
```

```python
# 把RGB三通道拆開顯示
import numpy as np

plt.figure(figsize = (15,15))
img = cv2.imread('.\D00_lena.png', cv2.IMREAD_COLOR)
for i,x in enumerate(['red', 'green', 'blue']):
    globals()[x] = np.zeros(img.shape[:], dtype = "uint8")
    globals()[x][:, :, i] = cv2.split(img)[-(i-1)+1]
    plt.subplot(1, 3, i+1, title = x)
    plt.imshow(globals()[x],cmap = 'hsv')
    plt.axis('off')
plt.show()
```
