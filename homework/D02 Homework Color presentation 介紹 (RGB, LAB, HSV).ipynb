- 改變圖片的 color space (`HLS`, `LAB`) 並呈現

```python
import cv2
import matplotlib.pyplot as plt

for i in {'brg' : 0, 'hsv' : cv2.COLOR_BGR2HSV, 'hls' : cv2.COLOR_BGR2HLS, 'lab' : cv2.COLOR_BGR2LAB}.items():
    img = cv2.imread('.\D00_lena.png')
    globals()[i[0]] = cv2.cvtColor(img, i[1]) if i[1] else img
    
    print(i[0])
    plt.imshow(globals()[i[0]][:, :, [2, 1, 0]],cmap = 'brg')
    plt.show()

# # opencv 的開啟圖片
# while True:
#     cv2.imshow('bgr', img)
#     cv2.imshow('hsv', hsv)
#     cv2.imshow('hls', hls)
#     cv2.imshow('lab', lab)
     
#     k = cv2.waitKey(0)
#     if k == 27:
#         cv2.destroyAllWindows()
#         break
```
